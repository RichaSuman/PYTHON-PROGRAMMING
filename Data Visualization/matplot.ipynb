Basics of Matplotlib
    To plot a graph from pandas Dataframe we need to convert it to Numpy Arrays


```python
%matplotlib inline
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filename = '/home/richa-mishra/Desktop/Python Programming/Data Visualization/Ex_Files_Python_for_Data_Vis/Exercise Files/Pandas/data/car_financing.csv'
df = pd.read_csv(filename)
```

```python
df.head()
```

```python
# to check the NaN Data as not easy to plot data with NaNs
df.info()
```

```python
Interest_missing=df['Interest Paid'].isna()
df.loc[Interest_missing,'Interest Paid'] = 93.24 
```

```python
df.info()
```

```python
# Covert to Numpy Array
# To plot a graph from pandas Dataframe we need to convert it to Numpy Arrays
month_number = df.loc[:,'Month'].values
interest_paid = df.loc[:,'Interest Paid'].values
principal_paid = df.loc[:,'Principal Paid'].values
```

Change Marker Type or Colour

```python
plt.style.use('seaborn')
plt.plot(month_number,interest_paid,marker='.',markersize=10)
plt.plot(month_number,principal_paid,marker='.',markersize=10)
```

```python
# Change colour using C parameter accept string
plt.plot(month_number,interest_paid,c= 'k',marker='.',markersize=10)
plt.plot(month_number,principal_paid,c='b',marker='.',markersize=10)
```

```python
## Can also use hex string for colour
# '#000000' is for Black
# '#0000FF' is for Blue
plt.plot(month_number,interest_paid,c= '#000000',marker='.',markersize=10)
plt.plot(month_number,principal_paid,c='#0000FF',marker='.',markersize=10)
```

```python
## Also we can use RGB Tuples
# (0,0,0) for Black
# (0,0,1) for Blue
plt.plot(month_number,interest_paid,c= (0,0,0),marker='.',markersize=10)
plt.plot(month_number,principal_paid,c=(0,0,1),marker='.',markersize=10)
```

Matplotlib Syntax
    Matlab Style
        Feels like Matlab
    Object-Oriented Syntax
        Use in situation where we want more control over the figure
    Plots are often made with combination of both the style

```python
## MATLAB SYNTAX
plt.style.use('seaborn')
plt.plot(month_number,interest_paid,c='k')
plt.plot(month_number,principal_paid,c='b')
```

```python
## OBJECT Oriented Syntax
x,y =(3,9)
fig, axes = plt.subplots(nrows =1, ncols =1 )
axes.plot(month_number,interest_paid,c='k')
plt.plot(month_number,principal_paid,c='b')
```

Plot Titles , Labels and Limits

```python
## MATLAB Plot
plt.style.use('seaborn')
plt.plot(month_number,interest_paid,c='k')
plt.plot(month_number,principal_paid,c='b')
```

```python
## Set xlim and ylim
plt.style.use('seaborn')
plt.plot(month_number,interest_paid,c='k')
plt.plot(month_number,principal_paid,c='b')
plt.xlim(left=1, right=70)
plt.ylim(bottom=0,top=2000)
```

```python
## set xlabel and ylabel and title
plt.xlabel('Month')
plt.ylabel('Dollars')
plt.title('Interest and Principal Paid each month')
plt.style.use('seaborn')
plt.plot(month_number,interest_paid,c='k')
plt.plot(month_number,principal_paid,c='b')
```

```python
## changing Fontsize
plt.xlabel('Month', fontsize=22)
plt.ylabel('Dollars', fontsize=22)
plt.title('Interest and Principal Paid each month', fontsize=22)
plt.style.use('seaborn')
plt.plot(month_number,interest_paid,c='k')
plt.plot(month_number,principal_paid,c='b')
```

```python
## Changing tick size
plt.xlabel('Month', fontsize=22)
plt.ylabel('Dollars', fontsize=22)
plt.title('Interest and Principal Paid each month', fontsize=22)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.style.use('seaborn')
plt.plot(month_number,interest_paid,c='k')
plt.plot(month_number,principal_paid,c='b')
```

```python
## OBJECT Oriented Syntax
x,y =(3,9)
fig, axes = plt.subplots(nrows =1, ncols =1 )
axes.plot(month_number,interest_paid,c='k')
axes.plot(month_number,principal_paid,c='b')
axes.set_xlim(left= 1, right = 100)
axes.set_ylim(bottom= 0 , top= 1500)
axes.set_xlabel('Month', fontsize=10)
axes.set_ylabel('Dollar', fontsize=10)
axes.set_title('Interest and Principal Paid each month', fontsize=10)
## ticks are different from matlab plot
axes.tick_params(axis='x',labelsize=10)
axes.tick_params(axis='y', labelsize=10)
```

```python
## Matlab Syntax
plt.figure(figsize=(10,5))
plt.plot(month_number,interest_paid,c='k',label='Interest')
plt.plot(month_number,principal_paid,c='b',label='Principal')
plt.xlim(left= 1, right = 100)
plt.ylim(bottom= 0 , top= 1500)
plt.xlabel('Month', fontsize=10)
plt.ylabel('Dollar', fontsize=10)
plt.title('Interest and Principal Paid each month', fontsize=10)
plt.legend(loc=(1.02,0),borderaxespad=0,fontsize=10)
## ticks are different from matlab plot
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.tight_layout() #  to adjust the plot in the given area
plt.savefig('/home/richa-mishra/Desktop/Python Programming/Data Visualization/Ex_Files_Python_for_Data_Vis/Exercise Files/Matplotlib/images/matplot.png',dpi=300)
# dpi is for the resolution
```
