```python
from sklearn.datasets import load_breast_cancer
```

```python
breast = load_breast_cancer()
dir(breast)
```

```python
print(breast.feature_names.shape)
```

```python

breast_data = load_breast_cancer().data
print(breast_data)
load_breast_cancer().data.shape
```

```python
breast_labels = load_breast_cancer().target
load_breast_cancer().target.shape
```

```python
import numpy as np

labels = np.reshape(breast_labels,(569,1))
#print(labels)
```

```python
# After reshaping the labels, you will concatenate the data and labels along the second axis, which means the final shape of the array will be 569 x 31.
final_breast_data = np.concatenate([breast_data,labels],axis=1)
print(final_breast_data.data)
```

```python
import pandas as pd
df=pd.DataFrame(final_breast_data)
#df
```

```python
features = breast.feature_names
features_labels = np.append(features,'label')
features_labels
```

```python
import pandas as pd
breast_dataset=pd.DataFrame(final_breast_data)
breast_dataset.columns=features_labels
breast_dataset
```

```python
import pandas as pd
breast_dataset=pd.DataFrame(data=final_breast_data,columns=features_labels)
breast_dataset
```

```python
breast_dataset.head()
```

```python
breast_dataset.loc[breast_dataset['label'] == 0, 'label'] = 'Benign'
breast_dataset.loc[breast_dataset['label'] == 1, 'label'] = 'Malignant'
#breast_dataset['label'].replace(0, 'Benign',inplace=True)
#breast_dataset['label'].replace(1, 'Malignant',inplace=True)

breast_dataset.shape
```

```python
from sklearn.preprocessing import StandardScaler
x= breast_dataset.loc[:,features].values
x=StandardScaler().fit_transform(x) # normalise the features
x
```

```python
# Let's check whether the normalized data has a mean of zero and a standard deviation of one.
np.mean(x),np.std(x)
```

```python
feat_cols = ['feature_'+str(i) for i in range(x.shape[1])] 
```

```python
normalised_breast_dataset = pd.DataFrame(data=x,columns=feat_cols)
normalised_breast_dataset.tail()
```

```python
from sklearn.decomposition import PCA
pca_breast = PCA(n_components=2)
principalComponents_breast = pca_breast.fit_transform(x) # normalize the data
```

```python
principal_breast_Df = pd.DataFrame(data = principalComponents_breast , columns = ['principal component 1', 'principal component 2'])
principal_breast_Df
```

```python
#print('Explained variation per principal component: {}'.format(pca_breast.explained_variance_ratio_))
print(f'Explained variation per principal component: {pca_breast.explained_variance_ratio_}')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure()
plt.figure(figsize=(15,15))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.title("Principal Component Analysis of Breast Cancer Dataset",fontsize=20)
targets = ['Benign', 'Malignant']
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = breast_dataset['label'] == target
    plt.scatter(principal_breast_Df.loc[indicesToKeep, 'principal component 1']
               , principal_breast_Df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})
```
