```python
%matplotlib inline
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Ex_Files_Python_for_Data_Vis/Exercise Files/Common Plot Applications/data/kingCountyHouseData.csv')
df.head()
```

```python
df['price'].head()
```

```python
df['price'].hist()
## Plot not readable
```

```python
df['price'].hist()
plt.xticks(rotation=90)
```

```python
## can also use seaborn style as all the plot styles have few default setting which can be benefitial
plt.style.use('seaborn')
df['price'].hist(bins=100)
```

```python
## mostly we are always interested in getting subset of the entire data set
price_filter = df.loc[:,'price'] <= 3000000
df.loc[price_filter,'price'].hist(bins=30)
```

```python
price_filter = df.loc[:,'price'] <= 3000000
df.loc[price_filter,'price'].hist(bins=30,
                                  edgecolor='black')
```
